Research Question # 1 : **Is there an association between education enrollment status and alcohol consumption?**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pip install ResearchPy
```

```python
import researchpy as rp
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

**Codebook 1: Enrollment Status**

Screenshot_3.png

```python
#Manage data for variable 1

df['Enrollment_Status'] = df['H4ED6']

df['Enrollment_Status'] = df['Enrollment_Status'].loc[df['Enrollment_Status']!='6']
df['Enrollment_Status'] = df['Enrollment_Status'].loc[df['Enrollment_Status']!='8']
df['Enrollment_Status'] = df['Enrollment_Status'].loc[df['Enrollment_Status']!=' ']

df['Enrollment_Status'].replace(['0','1'], ['NO','YES'], inplace = True)
df['Enrollment_Status'].value_counts()
```

```python
#Frequency Table

rp.summary_cat(df['Enrollment_Status'])
```

```python
#Bar Graph

plt.title("Are you currently enrolled in college, university, or school?")
df['Enrollment_Status'].value_counts(normalize = True).plot(kind ='bar', color = 'maroon', width = 0.2)
```

```python
#Pie Chart

plt.title("Are you currently enrolled in college, university, or school?")
df['Enrollment_Status'].value_counts().plot(kind ='pie', autopct='%1.1f%%')
```

Codebook 2: **Alcohol_Consumption**

Screenshot_6.png

```python
#Manage data for variable 2

df['Alcohol_Consumption'] = df['H4TO33']

df['Alcohol_Consumption'] = df['Alcohol_Consumption'].loc[df['Alcohol_Consumption']!=' ']
df['Alcohol_Consumption'] = df['Alcohol_Consumption'].loc[df['Alcohol_Consumption']!='6']
df['Alcohol_Consumption'] = df['Alcohol_Consumption'].loc[df['Alcohol_Consumption']!='8']

df['Alcohol_Consumption'].replace(['0','1'], ['NO','YES'], inplace = True)
```

```python
#Frequency Table

rp.summary_cat(df['Alcohol_Consumption'])
```

```python
#Bar Graph
plt.title("Have you had more than two alcoholic beverages?")
df['Alcohol_Consumption'].value_counts(normalize = True).plot(kind ='bar', color = 'green', width = 0.2)
```

```python
#Pie Chart
plt.title("Have you had more than two alcoholic beverages?")
df['Alcohol_Consumption'].value_counts().plot(kind ='pie', autopct='%1.1f%%')
```
