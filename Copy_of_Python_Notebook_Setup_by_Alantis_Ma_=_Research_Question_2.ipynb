Research Question # 2 : **Is there an association between the amount of fast foods eaten and time spent on the internet, in the past week?**




```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

**Codebook 1: Internet_Hours_Week**

Screenshot_4.png

```python
#Manage data for variable 1

df['Internet_Hours_Week'] = df['H4DA22']

df['Internet_Hours_Week']= pd.to_numeric(df['Internet_Hours_Week'],errors= 'coerce')
df['Internet_Hours_Week']= df['Internet_Hours_Week'].loc[df['Internet_Hours_Week'] < 996]
df['Internet_Hours_Week']= df['Internet_Hours_Week'].loc[df['Internet_Hours_Week'] != ' ']
```

```python
#Frequency Table

df['Internet_Hours_Week'].describe()
```

```python
#Boxplot

sns.boxplot(x = 'Internet_Hours_Week', data =df, color ='green')
```

```python
#Histogram

sns.histplot(x= 'Internet_Hours_Week', data = df, stat='percent', bins=14, kde=True)
```

**Codebook 2: Fast_foods_Week**

Screenshot_5.png

```python
#Manage data for variable 2

df['Fast_Foods_Week'] = df['H4GH8']

df['Fast_Foods_Week']= pd.to_numeric(df['Fast_Foods_Week'],errors= 'coerce')
df['Fast_Foods_Week']= df['Fast_Foods_Week'].loc[df['Fast_Foods_Week'] < 996]
df['Fast_Foods_Week']= df['Fast_Foods_Week'].loc[df['Fast_Foods_Week'] != ' ']
```

```python
#Frequency Table

df['Fast_Foods_Week'].describe()
```

```python
#Boxplot

sns.boxplot(x = 'Fast_Foods_Week', data =df, color ='purple')
```

```python
#Histogram

sns.histplot(x= 'Fast_Foods_Week', data = df, stat='percent', bins=14, kde=True)
```
